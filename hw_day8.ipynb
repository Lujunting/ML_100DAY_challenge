```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
data_path='C:/Users/User/Documents/'
```

```python
f_app_train=os.path.join(data_path,'application_train.csv')
app_train=pd.read_csv(f_app_train)
```

```python
app_train.head()
```

```python
# further to understand the detailof data
app_train.describe()
```

```python
# the three type of data collection
int_box=[]
float_box=[]
object_box=[]
for dtype, feature in zip(app_train.dtypes, app_train.columns):
    if dtype=='int64':
        int_box.append(feature)
    elif dtype=='float64':   
        float_box.append(feature)
    else:
        object_box.append(feature)
```

```python
app_train[int_box].describe()
app_train[int_box].mean()
app_train[int_box].std()
```

```python
app_train[float_box].describe()
app_train[float_box].mean()
app_train[float_box].std()
```

```python
app_train[object_box].describe()
```

```python
# plot Histogram
print('mean value:',app_train['NAME_INCOME_TYPE'].mean())
print('std value:',app_train['NAME_INCOME_TYPE'].std())
```
