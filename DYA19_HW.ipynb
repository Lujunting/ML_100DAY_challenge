```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read data
dir_data='C:/Users/User/Documents/application_train.csv'
f_app=os.path.join(dir_data)
f_train=pd.read_csv(f_app)
f_train.head()
```

```python
# 使用不同的house mode 類別繪製，用subplot排版
house_mode=f_train['HOUSETYPE_MODE'].unique()

nrows=len(house_mode)
ncols=nrows//2

plt.figure(figsize=(10,10))
for i in range(len(house_mode)):
    plt.subplot(nrows,ncols,i+1)
    
    f_train.loc[f_train['HOUSETYPE_MODE']==house_mode[i],'AMT_CREDIT'].hist()
    plt.title(str(house_mode[i]))
plt.show()
```
